```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
house = pd.read_csv('../housing.csv')
house
```

```python
X = house.drop('median_house_value', axis=1).copy()
y = house.median_house_value
```

```python
X_num = X.select_dtypes('float')
X_num
```

```python
X_cat = X.select_dtypes('object')
X_cat
```

```python
from sklearn.impute import SimpleImputer
```

```python
impute = SimpleImputer(strategy='median')
```

```python
impute.fit(X_num)
```

```python
impute.transform(X_num)
```

```python
X_num_impute = pd.DataFrame(impute.transform(X_num), columns=impute.get_feature_names_out())
X_num_impute
```

```python
X_num.isna().sum()
```

```python
X_num_impute.isna().sum()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale = StandardScaler()
```

```python
scale.fit(X_num_impute)
```

```python
scale.mean_
```

```python
scale.var_
```

```python
scale.transform(X_num_impute)
```

```python
X_num_tr = pd.DataFrame(scale.transform(X_num_impute), columns=scale.get_feature_names_out())
X_num_tr
```

```python
X_cat
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder()
```

```python
encoder.fit(X_cat)
```

```python
encoder.categories_
```

```python
encoder.get_feature_names_out()
```

```python
sparse_matrix = encoder.transform(X_cat)
sparse_matrix
```

```python
X_cat_tr = pd.DataFrame(sparse_matrix.toarray(), columns=encoder.get_feature_names_out())
X_cat_tr
```

```python
X_num_tr
```

```python
X_cat_tr
```

```python
X_final = pd.concat([X_num_tr, X_cat_tr], axis=1)
X_final
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_final, y, test_size=0.2, random_state=123)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression, SGDRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR

from sklearn.metrics import r2_score, mean_absolute_error
```

```python
models = [
    ('linear_regression', LinearRegression()),
    ('SGD', SGDRegressor()),
    ('Random_forest', RandomForestRegressor()),
    ('Tree', DecisionTreeRegressor()),
    ('neighbores', KNeighborsRegressor()),
    ('SVM', SVR())
]
```

```python
for name, model in models:
    model.fit(X_train, y_train)

    print(f"{name}")
    y_hat = model.predict(X_train)

    print(f"Training error: {mean_absolute_error(y_train, y_hat):.2f}")
    print(f"Training accuracy: {r2_score(y_train, y_hat):.2f}")

    print("_"*100)

    y_hat = model.predict(X_test)

    print(f"Testing error: {mean_absolute_error(y_test, y_hat):.2f}")
    print(f"Testing accuracy: {r2_score(y_test, y_hat):.2f}")
    print()
```
