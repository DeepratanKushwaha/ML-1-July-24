```python
def predict(m, X, c):
    y_hat = m * X + c
    return y_hat
```

```python
def MAE(y, y_hat):
    error = np.abs(y - y_hat).sum() / len(y)
    return error
```

```python
def corr_plot(X, y):
    r = np.corrcoef(height, weight)[0,1]
    plt.figure()
    plt.scatter(X, y, label=f"r ={r:.2f}")
    plt.legend()
    plt.grid()
    return plt.gca()
```

```python
ax = corr_plot(height, weight)

m = 0.2
c = 2

y_hat = predict(m, height, c)

error = MAE(weight, y_hat)

plt.plot(height, y_hat, label=f"error: {error}")
plt.legend()
plt.show()
```

```python
ax = corr_plot(height, weight)

m = 0.2
c = 2
y_hat = predict(m, height, c)
error = MAE(weight, y_hat)
plt.plot(height, y_hat, label=f"error: {error}")


m = 0.4
c = 4
y_hat = predict(m, height, c)
error = MAE(weight, y_hat)
plt.plot(height, y_hat, label=f"error: {error}")


m = 0.6
c = 5
y_hat = predict(m, height, c)
error = MAE(weight, y_hat)
plt.plot(height, y_hat, label=f"error: {error}")


plt.legend()
plt.show()
```

```python
for i in house:
    y = house['median_house_value']
    X = house[i]
    if i == 'median_house_value' or i == 'ocean_proximity':
        continue
        
    r = np.corrcoef(X,y)[0,1]
    plt.scatter(X, y, label=f"r = {r}")
    
    plt.title(f"{i} vs median_house_value")
    plt.xlabel(i)
    plt.ylabel('Median_house_value')
    plt.legend()
    plt.grid()
    plt.show()
```
