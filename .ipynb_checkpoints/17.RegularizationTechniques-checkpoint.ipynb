```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mpg = sns.load_dataset('mpg')
```

```python
mpg.isna().sum()
```

```python
mpg.dropna(inplace=True)
```

```python
X = mpg[['displacement', 'horsepower', 'acceleration', 'weight']]
y = mpg.mpg
```

```python
for feature in X:
    r = np.corrcoef(X[feature], y)[0, 1]
    sns.scatterplot(x=X[feature], y=y, label=f"r = {r:.2f}")
    plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
y_pred = model.predict(X)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
r2_score(y, y_pred)
```

```python
np.sqrt(mean_squared_error(y, y_pred))         # rmse
```

```python
for feature in X:
    r = np.corrcoef(X[feature], y)[0, 1]
    sns.scatterplot(x=X[feature], y=y, label=f"r = {r:.2f}")
    plt.scatter(X[feature], y_pred, color='r', alpha=0.3)
    plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
```

```python
model = Pipeline([
    ('poly', PolynomialFeatures(degree=2, include_bias=False)),
    ('LR', LinearRegression())
])
```

```python
model
```

```python
model.fit(X, y)
```

```python
y_pred = model.predict(X)
y_pred
```

```python
print(f"r2 score: {r2_score(y, y_pred)}")
print(f"Error: {np.sqrt(mean_squared_error(y, y_pred))}")
for feature in X:
    r = np.corrcoef(X[feature], y)[0, 1]
    sns.scatterplot(x=X[feature], y=y, label=f"r = {r:.2f}")
    plt.scatter(X[feature], y_pred, color='r', alpha=0.3)
    plt.show()
```

```python
model = Pipeline([
    ('poly', PolynomialFeatures(degree=3, include_bias=False)),
    ('LR', LinearRegression())
])


model.fit(X, y)
y_pred = model.predict(X)

print(f"r2 score: {r2_score(y, y_pred)}")
print(f"Error: {np.sqrt(mean_squared_error(y, y_pred))}")
for feature in X:
    r = np.corrcoef(X[feature], y)[0, 1]
    sns.scatterplot(x=X[feature], y=y, label=f"r = {r:.2f}")
    plt.scatter(X[feature], y_pred, color='r', alpha=0.3)
    plt.show()
```

```python
model = Pipeline([
    ('poly', PolynomialFeatures(degree=7, include_bias=False)),
    ('LR', LinearRegression())
])


model.fit(X, y)
y_pred = model.predict(X)

print(f"r2 score: {r2_score(y, y_pred)}")
print(f"Error: {np.sqrt(mean_squared_error(y, y_pred))}")
for feature in X:
    r = np.corrcoef(X[feature], y)[0, 1]
    sns.scatterplot(x=X[feature], y=y, label=f"r = {r:.2f}")
    plt.scatter(X[feature], y_pred, color='r', alpha=0.3)
    plt.show()
```
