```python
data
```

```python
ordinal = OrdinalEncoder()
```

## one hot encoding


                nominal data

```python
data = pd.DataFrame({
    'gender': np.random.choice(['male', 'female'], 10, True),
    'food':np.random.choice(['south', 'north', 'east', 'west'], 10, True)
})
data
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder()
```

```python
encoder.fit(data)
```

```python
encoder.categories_
```

```python
encoder.get_feature_names_out()
```

```python
encoder.transform(data)
```

```python
sparse_matrix = encoder.transform(data)
sparse_matrix
```

```python
sparse_matrix.toarray()
```

```python
pd.DataFrame(sparse_matrix.toarray(), columns=encoder.get_feature_names_out())
```
