```python
from nltk.corpus import stopwords
```

```python
stopword = stopwords.words("english")
stopword
```

```python
def process_text(corpus):
    i = 0
    while i < len(corpus):
        msg = corpus[i].lower()
        sentance = []
        for word in msg.split():
            if word not in stopword:
                sentance.append(word)
        msg = " ".join(sentance)
        corpus[i] = msg
        i += 1
    return " ".join(corpus)
        
```

```python
s = process_text(corpus)
s
```
