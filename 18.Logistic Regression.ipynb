```python
import latexify
```

```python
def accuracy_score(y, y_pred):
    return (y == y_pred).sum() / len(y)
```

```python
@latexify.function
def accuracy_score(y, y_pred):
    return sum(y == y_pred) / len(y)
```

```python
less_hours = np.random.normal(3, 1.5, 50)
more_hours = np.random.normal(8, 1.5, 50)
```

```python
less_hours
```

```python
more_hours
```

```python
hours = np.concatenate((less_hours, more_hours))
hours
```

```python
results = np.concatenate((np.zeros(50), np.ones(50)))
results
```

```python
sns.regplot(x=hours, y=results)
```

```python
plt.figure(dpi=120)
sns.regplot(x=hours, y=results)
plt.xlabel('hours')
plt.ylabel('result')
plt.show()
```

```python
plt.figure(dpi=120)
sns.regplot(x=hours, y=results)
plt.yticks([0, 1], ['fail', 'pass'])
plt.xlabel('hours')
plt.ylabel('result')
plt.show()
```

```python
from sklearn.linear_model import SGDRegressor
```

```python
model = SGDRegressor()
```

```python
X = hours.reshape(-1, 1)
y = results
```

```python
model.fit(X, y)
```

```python
y_pred = model.predict(X)
y_pred
```

```python
plt.figure(dpi=120)
plt.scatter(X,y)
plt.plot(X, y_pred)
plt.show()
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y, y_pred)
```

```python
def predict(y_pred, th=0.5):
    result = []
    for i in y_pred:
        if i < th:
            result.append(0.0)
        else:
            result.append(1.0)
    return result
```

```python
y_hat = predict(y_pred)
print(y_hat)
```

```python
y
```

```python
accuracy_score(y, y_hat)
```
