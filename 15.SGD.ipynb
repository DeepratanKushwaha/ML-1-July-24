```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
height = np.random.normal(160, 20, 1000)
bias = np.random.normal(10, 3, 1000)
weight = 0.3 * height + bias
```

```python
bias.mean()
```

```python
plt.figure(dpi=120)
sns.scatterplot(x=height, y=weight, label='Actual')
plt.show()
```

```python
t0 = np.random.random() * 3
t1 = np.random.random()
```

```python
print(t0)
print(t1)
```

```python
y_pred = t1 * height + t0
```

```python
from sklearn.metrics import mean_squared_error
```

```python
plt.figure(dpi=120)
sns.scatterplot(x=height, y=weight, label='Actual')
plt.xlabel('Height')
plt.ylabel('weight')
plt.plot(height, y_pred, label='prediction')
print(f"Error = {mean_squared_error(weight, y_pred):.2f}")
plt.legend()
plt.show()
```

```python
grad_t0 = (y_pred - weight).sum() / len(weight)
grad_t1 = ((y_pred - weight) * height).sum() / len(weight)..
```

```python
grad_t0
```

```python
grad_t1
```

```python
t0_next = t0 - grad_t0
t1_next = t1 - grad_t1
```

```python
t0_next
```

```python
t1_next
```

```python
plt.figure(dpi=120)
sns.scatterplot(x=height, y=weight, label='Actual')
plt.xlabel('Height')
plt.ylabel('weight')
plt.plot(height, y_pred, label='prediction')
print(f"Error = {mean_squared_error(weight, y_pred):.2f}")

y_pred = t1_next * height + t0_next
plt.plot(height, y_pred, label='optimize')
print(f"Error = {mean_squared_error(weight, y_pred):.2f}")

plt.legend()
plt.show()
```

```python
eta = 0.01
plt.figure(dpi=120)
sns.scatterplot(x=height, y=weight, label='Actual')
plt.xlabel('Height')
plt.ylabel('weight')

y_pred = t1 * height + t0
plt.plot(height, y_pred, label='prediction')
print(f"Error = {mean_squared_error(weight, y_pred):.2f}")

grad_t1 = ((y_pred - weight) * height).sum() / len(weight)
grad_t0 = (y_pred - weight).sum() / len(weight)

t0_next = t0 - eta * grad_t0
t1_next = t1 - eta * grad_t1

y_pred = t1_next * height + t0_next
plt.plot(height, y_pred, label='optimize')
print(f"Error = {mean_squared_error(weight, y_pred):.2f}")

plt.legend()
plt.show()
```

```python
eta = 0.00001
max_iter = 1000

t0 = np.random.random() * 3
t1 = np.random.random()

plt.figure(dpi=120)
sns.scatterplot(x=height, y=weight, label='Actual')
plt.xlabel('Height')
plt.ylabel('weight')
y_pred = t1 * height + t0

for i in range(max_iter):
    grad_t1 = ((y_pred - weight) * height).sum() / len(weight)
    grad_t0 = (y_pred - weight).sum() / len(weight)

    t0 = t0 - eta * grad_t0
    t1 = t1 - eta * grad_t1
    
    y_pred = t1 * height + t0
    print(f"Error = {mean_squared_error(weight, y_pred):.2f}")
    plt.plot(height, y_pred, label=f"{i}")
    
# plt.legend()   
plt.show()  
```
